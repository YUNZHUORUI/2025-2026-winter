```python
'''
package 
'''
from sympy import symbols, Eq, sin, cos, atan2, sqrt, simplify
from IPython.display import display

```

```python
'''
initial setup
''' 
# functions
def DH_matrix(alpha, a, d, theta):
    from sympy import Matrix
    T = Matrix([[cos(theta), -sin(theta), 0, a],
                [sin(theta)*cos(alpha), cos(theta)*cos(alpha), -sin(alpha), -d*sin(alpha)],
                [sin(theta)*sin(alpha), cos(theta)*sin(alpha), cos(alpha), d*cos(alpha)],
                [0, 0, 0, 1]])
    return T
 
```

```python
## variables
L0, L1, L2, L3, L4 = symbols('L0 L1 L2 L3 L4')
d = symbols('d') 
theta1, theta2, theta3 = symbols('theta1 theta2 theta3')
```

```python
T01 = DH_matrix(0, 0, L0, theta1+atan2(L1, L2))
T12 = DH_matrix(0, sqrt(L1**2 + L2**2), d, -atan2(L1, L2))
T23 = DH_matrix(0, L3, 0, theta2)
T34 = DH_matrix(0, L4, 0, theta3 )

display(T01)
display(T12)
display(T23)
display(T34)

T04 = T01 @ T12 @ T23 @ T34
display(T04)

T04_simplified = simplify(T04)
display(T04_simplified)
```
