Let's write the equation of static force in manipulation

```python
import numpy as np
from numpy import cos, sin, deg2rad ,array, cross
import sympy 
from sympy import symbols, Matrix, simplify
```

```python
# Variable initialization
theta1 = np.deg2rad(45)    
theta2 = np.deg2rad(80)
l1 = 1.2  #m
l2 = 0.8 #m
m= 1 #kg
g = 9.81 #m/s^2
```

```python
# Define rotation matrix
def Rot_z(theta):
    R = Matrix([[cos(theta), -sin(theta), 0],
                [sin(theta), cos(theta), 0],
                [0, 0, 1]])
    return R
```

For force transform, ${^{i}f_i = {}^{i}_{i+1}R {}^{i+1}f_{i+1}}$, we want to find torque $\tau_1$ and $\tau_2$ in link 1,2

```python
f_2_in_2 = array([-m*g*sin(np.pi - theta1 - theta2), m*g*cos(np.pi - theta1 -theta2), 0])  # Force at frame 2
Tau_2_in_2 = np.cross(array([l2, 0, 0]), f_2_in_2)  # Torque at frame 2
R_1_2 = Rot_z(theta2)  # Rotation from frame 2 to frame 1
f_1_in_1 = R_1_2 @ f_2_in_2  # Force at frame 1 due to frame 2
tau_1_in_1 = R_1_2 @ Tau_2_in_2 + np.cross(array([l1, 0, 0]), f_1_in_1)  # Torque at frame 1 due to frame 2
print("Torque at joint 2:", Tau_2_in_2[2])
print("Torque at joint 1:", tau_1_in_1[2])
```
