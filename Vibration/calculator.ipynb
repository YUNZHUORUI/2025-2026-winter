```python
g = 9.81
Ts = 800
Tinf = 300
mu = 29.0e-6
rou = 0.637
k = 0.0427
beta = 1.83e-3
v = 45.5e-6
Pr = 0.708
D = 5e-3
c = 1042
Ks = 20
rous = 8000
Cs= 500
```

```python
Ra = (g*beta*(Ts-Tinf)*D**3/v**2 ) * c*mu/k
display(Ra)
```

```python
Nu = 2 + (0.589*Ra**0.25/(1+(0.492/Pr)**0.56)**0.44)
display(Nu)
```

```python
h = Nu * k/D
display(h)
from numpy import pi
Q = h*pi*D**2*(Ts-Tinf)
display(Q)
```

```python
U = (4/3) * pi * (D/2)**3
S = pi * D**2
L = U/S
display(L)
Bi = h*L/Ks
display(Bi)
tau = rous * Cs * U / (h*S)
display(tau)
```
