```python
from numpy import array, sqrt, pi, poly1d
from matplotlib.pyplot import figure, plot, axis, show, grid
```

```python
m = 3 # kg
k = 1400 # N/m
kom = k/m
```

```python
omega_n = sqrt(kom)
print('natural circular frequency = {:.1f} rad/s')
print('natural frequency = {:.2f} Hz')
c = 5 # Ns/m
c_c = 2*m*omega_n
# c = 0.707*c_c
```

```python
cp = array([1, c/m, k/m])
zeta = cp[1]/2/omega_n
print('damping ratio zeta = {:.3f}')
cp = poly1d(cp)
cp
```

```python
rots = cp.roots
rots
```

```python
# %matplotlib qt5
figure(1)
plot(rots.real,rots.imag,'*r')
grid(True)
# axis([-20,20,-20,20])
axis('equal')
show()
```
