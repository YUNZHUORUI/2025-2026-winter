Consider once again problem 2.20 that were done in the Nov 1 tutorial.  A counter-clockwise moment $M(t) = 20 \sin 12 t$ kNm is applied to the beam, as shown in the hand-drawn figure below. 



```python
import numpy as np
from numpy import arctan2, exp, sin, cos, cosh, sinh 
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
m = 1750 
c = 3500
k = 7.0e5
a = 1.25
b = 2.50
```

```python
omega_n = np.sqrt(k*a*a/m/b/b)
zeta = c/m/2/omega_n
omega = 12.0
t = np.linspace(0, 5, 800)
F = - 10E3* 1j * exp(1j*omega*t)
r = omega/omega_n
G = 1/(1 - r*r + 2j*zeta*r)
Theta = G*F/k
theta = Theta.real
M = 20 * sin(12.0*t)
```

```python
from numpy import rad2deg
from regex import T


plt.plot(t, rad2deg(theta))
plt.plot(t, M, 'r')
plt.xlabel('Time (s)')
plt.ylabel('Theta (rad)')
plt.title('Angular Displacement vs Time')
plt.grid()
plt.show()
```
