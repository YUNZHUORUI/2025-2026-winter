```python
import numpy as np
from numpy import sqrt, deg2rad, cos, sin, angle
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 
w = 1  # 
t = np.linspace(0, 4*np.pi, 1000)  # 04

a = 12 * sqrt(12**2+8**2)/sqrt(30**2+19**2)
phi = deg2rad(66.04)
y1 = 12 * cos(w * t)
y2 = a * cos(w * t - phi)  

# 
fig = plt.figure(figsize=(16, 8))

# 
ax1 = plt.subplot(1, 2, 1)
ax1.plot(t, y1, label=r'$12\cos(t)$', linewidth=2.5, color='blue')
ax1.plot(t, y2, label=r'$4.874\cos(t - 66.04^\circ)$', linewidth=2.5, color='red', linestyle='--')

ax1.grid(True, alpha=0.3, linestyle='--')
ax1.set_title('Waveform: cos(t) and 4.874cos(t-66.04)', fontsize=16, pad=20)
ax1.set_xlabel('time', fontsize=14)
ax1.set_ylabel('amplitude', fontsize=14)

# 
ax1.set_xlim(0, 4*np.pi)

# x
ax1.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi, 5*np.pi/2, 3*np.pi, 7*np.pi/2, 4*np.pi])
ax1.set_xticklabels(['0', '/2', '', '3/2', '2', '5/2', '3', '7/2', '4'])

# 
ax1.legend(fontsize=12, loc='upper right')

# 
ax2 = plt.subplot(1, 2, 2)
ax2.set_aspect('equal', 'box')
ax2.grid(True, alpha=0.3, linestyle='--')

# 
phasor1 = np.array([12, 0])  # 120
phasor2 = np.array([a * cos(-phi), a * sin(-phi)])  # a-

# 
ax2.axhline(y=0, color='black', linewidth=0.5)
ax2.axvline(x=0, color='black', linewidth=0.5)

# 
ax2.quiver(0, 0, phasor1[0], phasor1[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.005, headwidth=8, headlength=10, label=r'$12\angle0^\circ$')
ax2.quiver(0, 0, phasor2[0], phasor2[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.005, headwidth=8, headlength=10, label=r'$4.874\angle-66.04^\circ$')

# 
ax2.text(phasor1[0]/2, phasor1[1]/2 - 0.5, '12', fontsize=12, color='blue', ha='center', va='center')
ax2.text(phasor2[0]/2, phasor2[1]/2, f'{a:.3f}', fontsize=12, color='red', ha='center', va='center')

# 
theta = np.linspace(0, -phi, 100)
radius = max(12, a) * 0.3
x_arc = radius * np.cos(theta)
y_arc = radius * np.sin(theta)
ax2.plot(x_arc, y_arc, 'k-', linewidth=1.5)

# 
angle_text_x = radius * 0.7 * np.cos(-phi/2)
angle_text_y = radius * 0.7 * np.sin(-phi/2)
ax2.text(angle_text_x, angle_text_y, r'$66.04^\circ$', fontsize=12, ha='center', va='center')

# 
max_val = max(12, a) * 1.2
ax2.set_xlim(-max_val, max_val)
ax2.set_ylim(-max_val, max_val)

# 
theta_circle = np.linspace(0, 2*np.pi, 100)
circle_radius = 1
ax2.plot(circle_radius * np.cos(theta_circle), circle_radius * np.sin(theta_circle), 
         'gray', linestyle=':', linewidth=1, alpha=0.5, label='Unit Circle')

ax2.set_title('Phasor Diagram', fontsize=16, pad=20)
ax2.set_xlabel('Real', fontsize=14)
ax2.set_ylabel('Imaginary', fontsize=14)
ax2.legend(fontsize=12, loc='upper right')

# 
plt.tight_layout()
plt.show()

# 
print(f"Phasor 1: 120")
print(f"Phasor 2: {a:.3f}{-np.degrees(phi):.2f}")
print(f"Magnitude of Phasor 2: {np.sqrt(phasor2[0]**2 + phasor2[1]**2):.3f}")
print(f"Phase of Phasor 2: {np.degrees(np.arctan2(phasor2[1], phasor2[0])):.2f}")
```

```python
import numpy as np
from numpy import sqrt, deg2rad, cos, pi, sin
import matplotlib.pyplot as plt

# 
w = 200 * pi  # 
T = 2 * pi / w  # 

# 2-3
num_periods = 2  # 2
t = np.linspace(0, num_periods * T, 1000)  # 

# 
a = 20 * 1 / sqrt(0.9842**2 + 0.37699**2)
phi_0 = deg2rad(15)  # 
phi = deg2rad(-5.959)  # 

y1 = 20 * cos(w * t + phi_0)  # 
y2 = a * cos(w * t + phi)     # 

# 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# ====================  ====================
ax1.plot(t, y1, label=r'Input: $20\cos(200\pi t + 15^\circ)$', linewidth=2, color='blue', alpha=0.8)
ax1.plot(t, y2, label=r'Output: $18.98\cos(200\pi t - 5.959^\circ)$', linewidth=2, color='red', alpha=0.8, linestyle='--')

ax1.grid(True, alpha=0.3, linestyle='--')
ax1.set_title('Voltage Waveforms Comparison', fontsize=16, pad=20)
ax1.set_xlabel('Time (s)', fontsize=14)
ax1.set_ylabel('Amplitude (V)', fontsize=14)
ax1.set_xlim(0, num_periods * T)

# x
tick_positions = [i * T for i in range(num_periods + 1)]
tick_labels = [f'{i}T' if i > 0 else '0' for i in range(num_periods + 1)]
ax1.set_xticks(tick_positions)
ax1.set_xticklabels(tick_labels)

# 
for i in range(1, num_periods):
    ax1.axvline(x=i*T, color='gray', linestyle=':', alpha=0.5, linewidth=1)

ax1.legend(fontsize=12, loc='upper right')

# ====================  ====================
ax2.set_aspect('equal', 'box')
ax2.grid(True, alpha=0.3, linestyle='--')

# 
ax2.axhline(y=0, color='black', linewidth=0.5)
ax2.axvline(x=0, color='black', linewidth=0.5)

# 
# 2015 ()
phasor_input = np.array([20 * cos(phi_0), 20 * sin(phi_0)])
# a-5.959 ()
phasor_output = np.array([a * cos(phi), a * sin(phi)])

# 
ax2.quiver(0, 0, phasor_input[0], phasor_input[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.005, headwidth=8, headlength=10, 
           label=r'Input: $20\angle15^\circ$', alpha=0.8)

# 
ax2.quiver(0, 0, phasor_output[0], phasor_output[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.005, headwidth=8, headlength=10,
           label=r'Output: $18.98\angle-5.959^\circ$', alpha=0.8)

# 
ax2.text(phasor_input[0]/2, phasor_input[1]/2 + 1, '20', 
         fontsize=12, color='blue', ha='center', va='center', fontweight='bold')
ax2.text(phasor_output[0]/2, phasor_output[1]/2 - 1, f'{a:.2f}', 
         fontsize=12, color='red', ha='center', va='center', fontweight='bold')

# 
ax2.plot([-25, 25], [0, 0], 'k--', alpha=0.3, linewidth=0.8)
ax2.plot([0, 0], [-25, 25], 'k--', alpha=0.3, linewidth=0.8)

# 
# 15
theta_input = np.linspace(0, phi_0, 100)
radius_input = 7
x_arc_input = radius_input * np.cos(theta_input)
y_arc_input = radius_input * np.sin(theta_input)
ax2.plot(x_arc_input, y_arc_input, 'b-', linewidth=1.5, alpha=0.7)
ax2.text(radius_input*0.7*np.cos(phi_0/2), radius_input*0.7*np.sin(phi_0/2), 
         r'$15^\circ$', fontsize=11, color='blue', ha='center', va='center')

# -5.959
theta_output = np.linspace(0, -phi, 100)
radius_output = 9
x_arc_output = radius_output * np.cos(theta_output)
y_arc_output = radius_output * np.sin(theta_output)
ax2.plot(x_arc_output, y_arc_output, 'r-', linewidth=1.5, alpha=0.7)
ax2.text(radius_output*0.7*np.cos(-phi/2), radius_output*0.7*np.sin(-phi/2), 
         r'$-5.959^\circ$', fontsize=11, color='red', ha='center', va='center')

# 
phase_diff = phi_0 - (-phi)  # 15 - (-5.959) = 20.959
theta_diff = np.linspace(-phi, phi_0, 100)
radius_diff = 11
x_arc_diff = radius_diff * np.cos(theta_diff)
y_arc_diff = radius_diff * np.sin(theta_diff)
ax2.plot(x_arc_diff, y_arc_diff, 'g-', linewidth=2, alpha=0.8)
ax2.text(radius_diff*0.8*np.cos(phase_diff/4), radius_diff*0.8*np.sin(phase_diff/4), 
         r'$\Delta\phi = 20.959^\circ$', fontsize=11, color='green', 
         ha='center', va='center', fontweight='bold')

# 
max_val = max(20, a) * 1.3
ax2.set_xlim(-max_val, max_val)
ax2.set_ylim(-max_val, max_val)

ax2.set_title('Phasor Diagram', fontsize=16, pad=20)
ax2.set_xlabel('Real (V)', fontsize=14)
ax2.set_ylabel('Imaginary (V)', fontsize=14)
ax2.legend(fontsize=12, loc='upper right')

# 
circle_theta = np.linspace(0, 2*np.pi, 100)
circle_radius = 1
ax2.plot(circle_radius * np.cos(circle_theta), circle_radius * np.sin(circle_theta), 
         'gray', linestyle=':', linewidth=1, alpha=0.3, label='Unit Circle')

# 
plt.tight_layout()

# 
print("=" * 50)
print("PHASOR INFORMATION")
print("=" * 50)
print(f"Input Phasor: 20{np.degrees(phi_0):.2f}")
print(f"Output Phasor: {a:.4f}{np.degrees(-phi):.4f}")
print(f"Phase Difference: {np.degrees(phi_0 - (-phi)):.4f}")
print(f"Output Magnitude: {a:.4f} V")
print(f"Input Magnitude: 20 V")
print(f"Gain: {a/20:.4f} (or {20*np.log10(a/20):.2f} dB)")
print("=" * 50)

plt.show()
```
