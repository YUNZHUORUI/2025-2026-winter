Experiment data:

|$ U(V)$ | 1.5 $V$ | 1.8 $V$ | 2.1 $V$ | 2.18 $V$ | 2.7 $V$ | 
|:--|:--|:--|:--|:--|:--|
|$I_B (\mu A)$| 178.8|161.8 | 143.7| 138|
|$I_C (\mu A)$|47727.3|43636.4|39545.45|38454.54|
|$\beta$ |266.93|269.69|275.19|256.92|

$$ I_C = \frac{5-U}{R_C} (\mu A)$$

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi, array
I_B  = array([178.8,161.8,143.7,138])
I_C = array([47727.3,43636.4,39545.45,38454.54])

plt.figure(figsize=(8, 6))
plt.plot(I_B, I_C, marker='o', linewidth=2, markersize=8, color='b', label='I_C vs I_B')

# 
plt.title('I_B vs I_C Characteristic Curve', fontsize=12)
plt.xlabel('I_B (A)', fontsize=12)
plt.ylabel('I_C (A)', fontsize=12)

# 
plt.grid(True, linestyle='--', alpha=0.6)

# 
plt.legend(fontsize=11)

# 
plt.xlim(min(I_B)-5, max(I_B)+5)
plt.ylim(min(I_C)-2000, max(I_C)+2000)

# 
plt.tight_layout()
plt.show()
```
